Implementation of Gradient Descent with variation in learning rate

```
#Importing the libraries
import random 
import math
import numpy as np
```

```
#Defining a function to optimize
def f(x):
  return x**2+5*x+6
```

```
#Defining the derivative 
def der_f(x):
  return 2*x+5
```

```
# defining a gradient descent function
def gradientdescentunivariate():
  x0=random.random() # Random float number b/w 0 and 1
  i=0 #iteration
  eta=0.01 # Learning rate
  xi=x0
  df_dx=der_f(xi)
  while(df_dx-0.0)>0.001: #threshold =0.001
    print(i)
    print(xi)
    print(df_dx)
    print('***********')
    df_dx=der_f(xi) # Calculating derivative
    xi=xi-eta*df_dx #updating the equation
    i+=1
  return xi
```

```
gradientdescentunivariate()
```

When Learning Rate=0.01 then no of Iterations =424 ,here we had found the minima of function.

Now let's increase learning rate to 0.05.
